```python
import pandas as pd
import matplotlib.pylab as plt
pressure = pd.read_csv('pressures.txt')
plt.plot(pressure,color = 'orange')
plt.xlabel('Timestep')
plt.ylabel('Pressure')
plt.title('Pressure over Time')
plt.savefig('pressure.png')
```

```python
import numpy as np
pressure = pd.read_csv('pressures.txt')
average_pressure = np.mean(pressure)
print(average_pressure)
temperature = pd.read_csv('temps.txt')
average_temperature = np.mean(temperature)
print(average_temperature)
```
